```python
file = open('testfile2.txt','r') 

text = file.read()
valeurs_string = text.split(',')[:-2]

valeurs = []

for i in valeurs_string:
    if i != ' ':
        if int(i) < 500:
            if int(i) > 4:
                valeurs.append(int(i))

import matplotlib.pyplot as plt
plt.plot(valeurs[1:][::5000])
plt.show()
```

```python
import numpy as np

signal = np.array(valeurs)
transform = list(abs(np.fft.fft(signal)))

plt.plot(transform[1:(len(transform)/2)])
plt.show()
```

```python
from scipy import fftpack

signal = np.array(valeurs[::2][1:])

# fréquence d’échantillonnage en Hz
fe = 8000
# durée en seconde
T = 10
# Nombre de point :
N = len(signal) + 1
# Array temporel :
t = np.arange(1.,N)/fe

# signal fréquentiel : on divise par la taille du vecteur pour normaliser la fft
fourier = fftpack.fft(signal)/np.size(signal)
# axe fréquentiel:
axe_f = np.arange(0.,N-2)*fe/N

# On plot
plt.figure()
plt.subplot(121)
plt.plot(t,signal,'-')
plt.xlabel('axe temporel, en seconde')
plt.subplot(122)
plt.plot(axe_f,np.abs(fourier[1:]))
plt.xlabel('axe frequentiels en Hertz')
plt.show()

# plot signal pure
a = list(np.abs(fourier[1:]))[:4000][::10]
plt.plot(a)
plt.show()
```
